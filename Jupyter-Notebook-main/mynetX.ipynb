```python
import networkx as nx
from matplotlib import pyplot as plt
```

```python
G = nx.Graph()
print(G.nodes())
print(G.edges())

print(type(G.nodes()))
print(type(G.edges()))
```

```python
G = nx.Graph()

G.add_node('A')
G.add_node('B')

G.add_nodes_from(['C','D','E','F','G'])

print(G.nodes())
print(G.edges())

print(type(G.nodes()))
print(type(G.edges()))
```

```python
G = nx.Graph()

G.add_node('A')
G.add_node('B')

G.add_nodes_from(['C','D','E','F','G'])

print(G.nodes())
print(G.edges())
nx.draw(G,with_labels=True,node_color='red')
```

```python
G = nx.Graph()

G.add_node('A')
G.add_node('B')

G.add_nodes_from(['C','D','E','F','G'])

G.add_edges_from([('A','C'),('C','F'),('F','G'),('G','D'),('D','B'),('B','F')])

print(G.nodes())
print(G.edges())
nx.draw(G,with_labels=True,node_color='red')
```

```python
G = nx.Graph()

G.add_node('A')
G.add_node('B')

G.add_nodes_from(['C','D','E','F','G'])

G.add_edges_from([('A','C'),('C','F'),('F','G'),('G','D'),('D','B'),('B','F')])

colors=['g','m','r','m','b','g','m']
size=[1000,400,200,700,500,800,50]
ed_col=['green','magenta','red','yellow','blue','gray','red']

print(G.nodes())
print(G.edges())
nx.draw(G,with_labels=True,node_color=colors,node_size=size,edge_color=ed_col)
plt.title('CITIES CONNECTION')
plt.savefig('graph.png')
```

```python
G = nx.Graph()
G.add_edge('a','b',weight=0.1)
G.add_edge('b','c',weight=1.5)
G.add_edge('a','c',weight=1.0)
G.add_edge('c','d',weight=2.2)

nx.draw(G,width_labels=True,node_color='red')
print(nx.shortest)
```
