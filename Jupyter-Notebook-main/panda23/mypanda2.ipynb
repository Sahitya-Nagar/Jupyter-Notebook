```python
student = {
    'first' : ['yash','manav','poojan','arpit','juzer','ishan','shubham','parth','naughty','kush'],
    'last' : ['patel','shah','patel','shah','doshi','doshi','patel','patel','patel','patel'],
    'email' : ['yash@gmail.com','manav@gmail.com','poojan@gmail.com','arpit@gmail.com','juzer@gmail.com','ishan@gmail.com','shubham@gmail.com','parth@gmail.com','kush@gmail.com','abhishek@gmail.com'],
    'college' : ['changa','svit','changa','iti','svit','nirma','svit','changa','svit','changa'],
    'per' : [90,80,70,20,46,82,67,94,58,91],
    'grade' : ['aa','bb','cc','aa','aa','bb','dd','aa','cc','aa']
}
```

```python
df=pd.DataFrame(student)
df
```

```python
def toupper(x) :
    return x.upper()
```

```python
df['college']=df['college'].apply(toupper)
df
```

```python
# wap to change grade to 'dist' if per>=70 , to 'first class' if per>=60 , to 'second class' if per>=50 , otherwise 'fail'
def change(g):
    if g>=70 :
        return "Dist"
    elif g<=60 :
        return "First Class"
    elif g<=50 :
        return "Second Class"
    else:
        return "Fail"
```

```python
df['grade']=df['per'].apply(change)
df
```

```python
def place(c):
    if c=='CHANGA':
        return 'Charusat'
    elif c=='SVIT':
        return 'Rajupura'
    elif c=='ITI':
        return 'Tarsali'
    else:
        return 'Ahmedabad'
```

```python
df['city']=df['college'].apply(place)
df
```
